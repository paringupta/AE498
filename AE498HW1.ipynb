```python
import numpy as np
from scipy.optimize import minimize

# -----------------------------
# Constants
# -----------------------------
AU_KM = 149597870.7  # km per AU

# -----------------------------
# Orbital Elements (01/01/2026)
# -----------------------------
# Earth
a1 = 1.4765067E+08      # km
e1 = 9.1669995E-03
i1 = np.deg2rad(4.2422693E-03)
w1 = np.deg2rad(6.64375167E+01)
O1 = np.deg2rad(1.4760836E+01)

# Apophis
a2 = 1.3793939E+08
e2 = 1.9097084E-01
i2 = np.deg2rad(3.3356539E+00)
w2 = np.deg2rad(1.2919949E+02)
O2 = np.deg2rad(2.0381969E+02)


# -----------------------------
# Rotation Matrices
# -----------------------------
def R3(theta):
    return np.array([
        [np.cos(theta), np.sin(theta), 0],
        [-np.sin(theta),  np.cos(theta), 0],
        [0, 0, 1]
    ])

def R1(theta):
    return np.array([
        [1, 0, 0],
        [0, np.cos(theta), np.sin(theta)],
        [0, -np.sin(theta),  np.cos(theta)]
    ])


# -----------------------------
# Position from Keplerian Elements
# -----------------------------
def position(a, e, i, O, w, f):
    r = a * (1 - e**2) / (1 + e*np.cos(f))

    r_pf = np.array([
        r*np.cos(f),
        r*np.sin(f),
        0
    ])

    Q = R3(O) @ R1(i) @ R3(w)
    return Q @ r_pf


# -----------------------------
# Distance Function
# -----------------------------
def distance(vars):
    f1, f2 = vars
    r1 = position(a1, e1, i1, O1, w1, f1)
    r2 = position(a2, e2, i2, O2, w2, f2)
    return np.linalg.norm(r1 - r2)


# ---------------------------------
# Coarse Grid Search
# ---------------------------------

N = 200  # grid resolution
f_vals = np.linspace(0, 2*np.pi, N)

min_dist = 1e20
best_pair = (0, 0)

for f1 in f_vals:
    for f2 in f_vals:
        r1 = position(a1, e1, i1, O1, w1, f1)
        r2 = position(a2, e2, i2, O2, w2, f2)
        d = np.linalg.norm(r1 - r2)
        if d < min_dist:
            min_dist = d
            best_pair = (f1, f2)

#print("Coarse MOID (km):", min_dist)
#print("Coarse f1, f2 (deg):",
      #np.rad2deg(best_pair[0]),
      #np.rad2deg(best_pair[1]))

# ---------------------------------
# Refine With Local Minimization
# ---------------------------------

result = minimize(distance, best_pair, method='Nelder-Mead')

f1_moid = result.x[0] % (2*np.pi)
f2_moid = result.x[1] % (2*np.pi)
moid_km = result.fun
moid_au = moid_km / AU_KM

print("MOID (km):", moid_km)
print("MOID (AU):", moid_au)
print("f1 Earth (deg):", np.rad2deg(f1_moid))
print("f2 Apophis (deg):", np.rad2deg(f2_moid))

```
